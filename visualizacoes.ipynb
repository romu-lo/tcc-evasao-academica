### Funções

```python
def get_ano_ultima_geracao(df: pd.DataFrame):
    return int(df[df['NUMERO_GERACAO'] == df['NUMERO_GERACAO'].max()]['DT_ANO_INGRES'].unique()[0])
```

```python
def get_df_geracoes_completas(df: pd.DataFrame):
    return df[df['GERACAO_COMPLETA'] == True]

def get_df_geracoes_incompletas(df: pd.DataFrame):
    return df[df['GERACAO_COMPLETA'] == False]

def get_df_sem_matriculados(df: pd.DataFrame):
    return df[df['MULTI_ST_ALUNO'] != 'MATRICULADO']

def get_df_cancelados(df: pd.DataFrame):
    return df[df['MULTI_ST_ALUNO'] == 'CANCELAMENTO']

def get_df_evadidos(df: pd.DataFrame):
    return df[df['BOOL_ST_ALUNO'] == True]
```

```python
def get_pct_evasao(df: pd.DataFrame):
    return (df['BOOL_ST_ALUNO'].value_counts(normalize=True), 
            get_df_geracoes_completas(df)['BOOL_ST_ALUNO'].value_counts(normalize=True), 
            get_df_geracoes_incompletas(df)['BOOL_ST_ALUNO'].value_counts(normalize=True))
```

```python
def vis_evadidos_x_nao_evadidos(df: pd.DataFrame):
    fig, ax = plt.subplots(2, 2, figsize=(12, 8))

    count_bool_st = sns.countplot(data=df, x='BOOL_ST_ALUNO', ax=ax[0, 0])
    count_bool_st.bar_label(count_bool_st.containers[0])
    ax[0, 0].set_title('Total')

    count_bool_st = sns.countplot(data=get_df_geracoes_completas(df), x='BOOL_ST_ALUNO', ax=ax[0, 1])
    count_bool_st.bar_label(count_bool_st.containers[0])
    ax[0, 1].set_title('Gerações Completas')

    count_bool_st = sns.countplot(data=get_df_geracoes_incompletas(df), x='BOOL_ST_ALUNO', ax=ax[1, 0])
    count_bool_st.bar_label(count_bool_st.containers[0])
    ax[1, 0].set_title('Gerações Incompletas')

    ax[1, 1].axis('off')

    plt.tight_layout()


def vis_filtro_evadidos_x_nao_evadidos(df: pd.DataFrame, filtro: str):
    fig, ax = plt.subplots(2, 2, figsize=(12, 8))

    count_bool_st = sns.countplot(data=df, x=filtro, hue='BOOL_ST_ALUNO', ax=ax[0, 0])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[0, 0].set_title('Total')

    count_bool_st = sns.countplot(data=get_df_geracoes_completas(df), x=filtro, hue='BOOL_ST_ALUNO', ax=ax[0, 1])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[0, 1].set_title('Gerações Completas')

    count_bool_st = sns.countplot(data=get_df_geracoes_incompletas(df), x=filtro, hue='BOOL_ST_ALUNO', ax=ax[1, 0])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[1, 0].set_title('Gerações Incompletas')

    ax[1, 1].axis('off')

    plt.tight_layout()
```

```python
def vis_situacao_multiclasse(df: pd.DataFrame):
    fig, ax = plt.subplots(2, 2, figsize=(14, 9))

    count_multi_st = sns.countplot(data=df, x='MULTI_ST_ALUNO', ax=ax[0, 0])
    count_multi_st.bar_label(count_multi_st.containers[0])
    ax[0, 0].set_title('Total')

    count_multi_st = sns.countplot(data=get_df_geracoes_completas(df), x='MULTI_ST_ALUNO', ax=ax[0, 1])
    count_multi_st.bar_label(count_multi_st.containers[0])
    ax[0, 1].set_title('Gerações Completas')

    count_multi_st = sns.countplot(data=get_df_geracoes_incompletas(df), x='MULTI_ST_ALUNO', ax=ax[1, 0])
    count_multi_st.bar_label(count_multi_st.containers[0])
    ax[1, 0].set_title('Gerações Incompletas')

    ax[1, 1].axis('off')

    plt.tight_layout()

def vis_filtro_situacao_multiclasse(df: pd.DataFrame, filtro: str):
    fig, ax = plt.subplots(2, 2, figsize=(12, 8))

    count_bool_st = sns.countplot(data=df, x=filtro, hue='MULTI_ST_ALUNO', ax=ax[0, 0])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[0, 0].set_title('Total')

    count_bool_st = sns.countplot(data=get_df_geracoes_completas(df), x=filtro, hue='MULTI_ST_ALUNO', ax=ax[0, 1])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[0, 1].set_title('Gerações Completas')

    count_bool_st = sns.countplot(data=get_df_geracoes_incompletas(df), x=filtro, hue='MULTI_ST_ALUNO', ax=ax[1, 0])
    for container in count_bool_st.containers:
        count_bool_st.bar_label(container)
    ax[1, 0].set_title('Gerações Incompletas')

    ax[1, 1].axis('off')

    plt.tight_layout()
```

```python
def vis_tempo_cursado(df: pd.DataFrame):
    fig, ax = plt.subplots(2, 2, figsize=(15, 9))

    hist_meses = sns.histplot(df['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[0, 0])
    ax[0, 0].set_title('Total')
    for container in hist_meses.containers:
        hist_meses.bar_label(container)

    hist_meses = sns.histplot(get_df_evadidos(df)['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[0, 1])
    for container in hist_meses.containers:
        hist_meses.bar_label(container)
    ax[0, 1].set_title('Apenas Evadidos')


    hist_meses = sns.histplot(get_df_geracoes_completas(df)['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[1, 0])
    for container in hist_meses.containers:
        hist_meses.bar_label(container)
    ax[1, 0].set_title('Gerações Completas')

    hist_meses = sns.histplot(get_df_geracoes_incompletas(df)['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[1, 1])
    for container in hist_meses.containers:
        hist_meses.bar_label(container)
    ax[1, 1].set_title('Gerações Incompletas')

    plt.tight_layout()
```
