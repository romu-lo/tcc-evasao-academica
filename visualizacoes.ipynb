```python
hist_meses = sns.histplot(df_sem_matriculados['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 200, 6)])
# hist_meses.bar_label(hist_meses.containers[0])
```

```python
hist_meses = sns.histplot(df_sem_matriculados['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)])

counts, bin_edges = np.histogram(df_sem_matriculados['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)])

for i in range(len(bin_edges)-1):
    x_position = (bin_edges[i] + bin_edges[i+1]) / 2
    y_position = counts[i] + counts.max() * 0.05
    plt.text(x_position, y_position, f'{bin_edges[i]:.1f}-{bin_edges[i+1]:.1f} ({counts[i]})', ha='center', rotation=90)

plt.show()
```

### Computação

```python
df_computacao = df_selecionado[df_selecionado['CD_CURSO'] == 26]
df_computacao_sem_matriculados = df_computacao[df_computacao['MULTI_ST_ALUNO'] != 'MATRICULADO']
df_computacao_geracoes_completas = df_computacao[df_computacao['GERACAO_COMPLETA'] == 1]
df_computacao_geracoes_incompletas = df_computacao[df_computacao['GERACAO_COMPLETA'] == 0]
```

```python
df_computacao.shape[0]
```

#### Evadidos x não evadidos

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))

count_bool_st = sns.countplot(data=df_computacao, x='BOOL_ST_ALUNO', ax=ax[0, 0])
count_bool_st.bar_label(count_bool_st.containers[0])
ax[0, 0].set_title('Total')

count_bool_st = sns.countplot(data=df_computacao_geracoes_completas, x='BOOL_ST_ALUNO', ax=ax[0, 1])
count_bool_st.bar_label(count_bool_st.containers[0])
ax[0, 1].set_title('Gerações Completas')

count_bool_st = sns.countplot(data=df_computacao_geracoes_incompletas, x='BOOL_ST_ALUNO', ax=ax[1, 0])
count_bool_st.bar_label(count_bool_st.containers[0])
ax[1, 0].set_title('Gerações Incompletas')

ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
df_computacao['BOOL_ST_ALUNO'].value_counts(normalize=True), df_computacao_geracoes_completas['BOOL_ST_ALUNO'].value_counts(normalize=True), df_computacao_geracoes_incompletas['BOOL_ST_ALUNO'].value_counts(normalize=True)
```

#### Situação multiclasse

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 9))

count_multi_st = sns.countplot(data=df_computacao, x='MULTI_ST_ALUNO', ax=ax[0, 0])
count_multi_st.bar_label(count_multi_st.containers[0])
ax[0, 0].set_title('Total')

count_multi_st = sns.countplot(data=df_computacao_geracoes_completas, x='MULTI_ST_ALUNO', ax=ax[0, 1])
count_multi_st.bar_label(count_multi_st.containers[0])
ax[0, 1].set_title('Gerações Completas')

count_multi_st = sns.countplot(data=df_computacao_geracoes_incompletas, x='MULTI_ST_ALUNO', ax=ax[1, 0])
count_multi_st.bar_label(count_multi_st.containers[0])
ax[1, 0].set_title('Gerações Incompletas')

ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

#### Tempo até última Situação

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 9))

hist_meses = sns.histplot(df_computacao['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[0, 0])
ax[0, 0].set_title('Total')

hist_meses = sns.histplot(df_computacao_sem_matriculados['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[0, 1])
ax[0, 1].set_title('Sem matriculados')

hist_meses = sns.histplot(df_computacao_geracoes_completas['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[1, 0])
ax[1, 0].set_title('Gerações Completas')

hist_meses = sns.histplot(df_computacao_geracoes_incompletas['MESES_ATE_ULTIMA_SITUACAO'], bins=[i for i in range(0, 125, 6)], ax=ax[1, 1])
ax[1, 1].set_title('Gerações Incompletas')

plt.tight_layout()
plt.show()
```
